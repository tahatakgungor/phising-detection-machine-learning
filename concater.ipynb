```python
import pandas as pd
import numpy as np
import os
```

**Dataset Folder's CSV Concater**

```python
file_names = os.listdir("./files/")
pds = []
```

```python
for file_name in file_names:
    pds.append(pd.read_csv('files/'+file_name))
```

```python
df = pd.concat(pds)
```

```python
df = df.reset_index().drop(['index'], axis=1)
```

```python
df.shape
```

```python
df.to_csv("dataset_uptodate.csv", index=None)
```
